```python
import numpy as np
import matplotlib.pyplot as plt

filelist=['./checkpoint/cifar10/resnet50/0.5.txt','./checkpoint/cifar10/resnet50/0.6.txt','./checkpoint/cifar10/resnet50/0.7.txt','./checkpoint/cifar10/resnet50/0.8.txt','./checkpoint/cifar10/resnet50/0.9.txt','./checkpoint/cifar10/resnet50/0.95.txt','./checkpoint/cifar10/resnet50/1.0.txt']
filelist2=['./downsample/cifar10/resnet50/0.5.txt','./downsample/cifar10/resnet50/0.6.txt','./downsample/cifar10/resnet50/0.7.txt','./downsample/cifar10/resnet50/0.8.txt','./downsample/cifar10/resnet50/0.9.txt','./downsample/cifar10/resnet50/0.95.txt','./downsample/cifar10/resnet50/1.0.txt']
```

```python
for fname in filelist:
    X=[]
    Y=[]
    Z=[]
    H=[]
for fname in filelist:
    data=np.loadtxt(fname,skiprows=1)
    X.append(float((fname.replace('.txt', '')).replace('./checkpoint/cifar10/resnet50/', '')))
    Y.append(max(data[:,4]))
for fname in filelist2:
    data=np.loadtxt(fname,skiprows=1)
    H.append(float((fname.replace('.txt', '')).replace('./downsample/cifar10/resnet50/', '')))
    Z.append(max(data[:,4]))
print(X,Y,Z)
plt.plot(X,Y,label='downsampe=weight')
plt.plot(H,Z,label='downsampe=1.0')
plt.xlabel('weight')
plt.ylabel('accuracy')
plt.title('Cifar10 Res50 Testing accuracy comparison')
plt.legend()    
plt.savefig('Cifar10 Res50 Testing accuracy comparison', dpi=500)
plt.show()
```
