```python
for fname in filelist:
    data=np.loadtxt(fname,skiprows=1)
    X=data[:,0]
    Y=data[:,3]
    plt.plot(X,Y,label= fname.replace('.txt', ''))
plt.xlabel('#epoch')
plt.ylabel('accuracy')
plt.title('Training accuracy')
plt.legend()    
plt.savefig('Training accuracy', dpi=500)
plt.show()
```
