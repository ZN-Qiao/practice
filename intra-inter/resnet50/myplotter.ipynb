```python
import numpy as np
import matplotlib.pyplot as plt

filelist=['0.01.txt','0.1.txt','1.0.txt','10.0.txt','100.0.txt']

for fname in filelist:
    data=np.loadtxt(fname,skiprows=1)
    X=data[:,0]
    Y=data[:,1]
    plt.plot(X,Y,label= fname.replace('.txt', ''))
plt.xlabel('#epoch')
plt.ylabel('loss')
plt.title('Training loss')
plt.legend()    
plt.savefig('Training loss', dpi=500)
plt.show()
```
